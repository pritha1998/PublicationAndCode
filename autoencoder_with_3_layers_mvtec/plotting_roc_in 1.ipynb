```python
import os
from sklearn import metrics
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
list_of_files = os.listdir(os.getcwd()) #list of files in the current directory
print(list_of_files)

for each_file in list_of_files:
    if each_file.endswith('pr.csv'):
      print (each_file)
      file = os.path.splitext(each_file)

      print(file)  # returns tuple of string

      print(file[0] + file[1])
      df=pd.read_csv(each_file)
      recall=list(df["recall"].values)
      precision=list(df["precision"].values)
      roc_auc = metrics.auc(recall, precision)
#display = metrics.RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc)
#display.plot()
      Str= file[0] 
      r_l = Str[:-3]  
      plt.plot(precision,recall,label= r_l +", PR-AUC="+str(round(roc_auc,4)))
      plt.legend(loc='lower right',bbox_to_anchor=(1.5, 0))
      plt.xlabel("Recall")
      plt.ylabel("Precision")
      plt.title("ROC")
        
plt.savefig('pr_combined.jpg',bbox_inches='tight')
plt.show()
```

```python
import platform,socket,re,uuid,json,psutil,logging

def getSystemInfo():
    try:
        info={}
        info['platform']=platform.system()
        info['platform-release']=platform.release()
        info['platform-version']=platform.version()
        info['architecture']=platform.machine()
        info['hostname']=socket.gethostname()
        info['ip-address']=socket.gethostbyname(socket.gethostname())
        info['mac-address']=':'.join(re.findall('..', '%012x' % uuid.getnode()))
        info['processor']=platform.processor()
        info['ram']=str(round(psutil.virtual_memory().total / (1024.0 **3)))+" GB"
        return json.dumps(info)
    except Exception as e:
        logging.exception(e)

json.loads(getSystemInfo())
```
